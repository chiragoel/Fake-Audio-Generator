# TTS



```python
import os
import sys
import torch
import time
import IPython
import numpy as np
import librosa


from TTS.tts.utils.generic_utils import setup_model
from TTS.utils.io import load_config
from TTS.tts.utils.text.symbols import symbols, phonemes
from TTS.utils.audio import AudioProcessor
from TTS.tts.utils.synthesis import synthesis
from TTS.vocoder.utils.generic_utils import setup_generator
from TTS.tts.utils.visual import plot_spectrogram

use_cuda = False
```



## Load TTS Model (Tacotron2)



```python
# model paths
TTS_MODEL = "data_tts/tts_model.pth.tar"
TTS_CONFIG = "data_tts/config.json"
VOCODER_MODEL = "data_tts/vocoder_model.pth.tar"
VOCODER_CONFIG = "data_tts/config_vocoder.json"
```

```python
# load configs
TTS_CONFIG = load_config(TTS_CONFIG)
VOCODER_CONFIG = load_config(VOCODER_CONFIG)
```

```python
# load the audio processor
TTS_CONFIG.audio['stats_path'] = 'data_tts/scale_stats_vocoder.npy'
ap = AudioProcessor(**TTS_CONFIG.audio)
```

```python
#LOAD TTS MODEL
# multi speaker 
speaker_id = None
speakers = []

# load the model
num_chars = len(phonemes) if TTS_CONFIG.use_phonemes else len(symbols)
model = setup_model(num_chars, len(speakers), TTS_CONFIG)

# load model state
cp =  torch.load(TTS_MODEL, map_location=torch.device('cpu'))

# load the model
model.load_state_dict(cp['model'])
if use_cuda:
    model.cuda()
model.eval()

# set model stepsize
if 'r' in cp:
    model.decoder.set_r(cp['r'])
```



## Load Vocoder (MB-MelGAN)



```python
# LOAD VOCODER MODEL
vocoder_model = setup_generator(VOCODER_CONFIG)
vocoder_model.load_state_dict(torch.load(VOCODER_MODEL, map_location="cpu")["model"])
vocoder_model.remove_weight_norm()
vocoder_model.inference_padding = 0
VOCODER_CONFIG.audio['stats_path'] = 'data_tts/scale_stats_vocoder.npy'
ap_vocoder = AudioProcessor(**VOCODER_CONFIG['audio'])    
if use_cuda:
    vocoder_model.cuda()
vocoder_model.eval()
```



## Predictions



```python
def tts(model, text, CONFIG, use_cuda, ap, use_gl, figures=True):
    t_1 = time.time()
    waveform, alignment, mel_spec, mel_postnet_spec, stop_tokens, inputs = synthesis(model, text, CONFIG, use_cuda, ap, speaker_id, style_wav=None,
                                                                             truncated=False, enable_eos_bos_chars=CONFIG.enable_eos_bos_chars)
    #mel_postnet_spec = ap._denormalize(mel_postnet_spec.T)
    if not use_gl:
        waveform = vocoder_model.inference(torch.FloatTensor(mel_postnet_spec.T).unsqueeze(0))
        waveform = waveform.flatten()
    if use_cuda:
        waveform = waveform.cpu()
    waveform = waveform.numpy()
    rtf = (time.time() - t_1) / (len(waveform) / ap.sample_rate)
    tps = (time.time() - t_1) / len(waveform)
    print(waveform.shape)
    print(" > Run-time: {}".format(time.time() - t_1))
    print(" > Real-time factor: {}".format(rtf))
    print(" > Time per step: {}".format(tps))
    IPython.display.display(IPython.display.Audio(waveform, rate=CONFIG.audio['sample_rate']))  
    return alignment, mel_postnet_spec, stop_tokens, waveform
```

##Input Sentence

```python
sentence = input('Enter the Sentence: ')
print(f'You entered {sentence}')
```



## Results



```python
align, spec, stop_tokens, wav = tts(model, sentence, TTS_CONFIG, use_cuda, ap, use_gl=False, figures=True)
```

```python
print(np.shape(spec))
plot_spectrogram(spec, ap,fig_size=(5,5))
```

```python
librosa.output.write_wav('/content/results/tts_model/tts_result.wav', wav, 22050, norm=False)
```

## Import Libraries

```python
import os
import argparse
import pickle
import glob
import random
import numpy as np
from tqdm import tqdm
import matplotlib.pyplot as plt

import librosa
from librosa.filters import mel as librosa_mel_fn

import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.utils.data.dataset import Dataset

SAMPLING_RATE = 22050  # Fixed sampling rate
```

## Preprocessing

```python
stats = np.load('/content/cyclegan/stats/lj_norm_stats.npz')
```

```python
def save_pickle(variable, fileName):
    with open(fileName, 'wb') as f:
        pickle.dump(variable, f)

def normalize_mel(wave_path,stats):
    wav_files = wave_path
    vocoder = torch.hub.load('descriptinc/melgan-neurips', 'load_melgan')

    wav_orig, _ = librosa.load(wav_files, sr=SAMPLING_RATE, mono=True)
    spec = vocoder(torch.tensor([wav_orig]))
    mel_spec = spec.cpu().detach().numpy()[0]


    mel_normalized = list()
    assert mel_spec.shape[-1] >= 64, f"Mel spectogram length must be greater than 64 frames, but was {mel_spec.shape[-1]}"
    app = (mel_spec - stats['mean']) / stats['std']
    mel_normalized.append(app)
    save_pickle(variable=mel_normalized,
                fileName="/content/cyclegan/stats/test_normalized.pickle")

    return mel_normalized
```

```python
m = normalize_mel('/content/results/tts_model/tts_result.wav',stats)
```

```python
m = np.asarray(m)
m = m.reshape((m.shape[1],m.shape[2]))
print(m.shape)
```

```python
plot_spectrogram(m.T,fig_size=(5,5))
```
